```python
import numpy as np
import matplotlib.pyplot as plt
```

Recordemos que `matplitlib.pyplot` lo importamos siempre como `plt` y que éste es el objeto *core* (o el más importante para nostros), ya que contiene todos los mètodos para crear todo tipo de gráficas.

Veamos qué pasa si usamos `plt.plot()` con una serie arbitraria de números

```python
plt.plot([1,2,3,4,10])
```

En este caso, como únicamente especificamos una lista de números, entonces `matplotlib` asumió que esta lista se refería a los valores $y$ de nuestra gráfica, y tomó los valores $x$ empezando en 0 y terminando en el número todal de elementos.

Notemos que después de ejecutar el código, aparece un texto que dice 

`[<matplotlib.lines.Line2D at 0x7f89415b7470>]`

Esto se debe a que `matplotlib` regresa el objeto de la gráfica **además** de dibujarla. Si quisiéramos mostrar la gráfica sin este texto, agregamos la línea `plt.show()`, y ejecutamos todo el código de un jalón.

```python
plt.plot([1,2,3,4,10])
plt.show()
```

Queríamos una gráfica de dispersión, pero el código produjo una gráfica simple de línea. Esto se debe a que éste es el comportamiendo default. 

#### ¿Cómo hacemos una gráfica de disersión?

Entendamos mejor los argumentos que espera la función `plt.plot()`. `plt.plot()` acepta **3** argumentos básicos en el siguiente orden: `(x, y, formato)`.

El formato es una combinación de `{color}{marker}{line}`. Por ejemplo, el formato `'go-'` está compuesto de 3 caracteres. 
- `g` se refiere a **'green'**
- `o` se refiere a **'dots'** u **o's**
- `-` se refiere a **solid line**

Entonces `'go-'` es equivalente a decir **'green dots with solid line'** o **puntos verdes con línea sólida** en español.

Por otro lado `'go'` sería **'green dots'** o **puntos verdes** en español. Y si tenemos puros puntos sin línea, entonces tenemos una gráfica de dispersión.

Veamos ambos ejemplos:

### `'go-'` $\rightarrow$ puntos verdes con línea sólida

```python
plt.plot([1,2,3,4,5], [1,2,3,4,10], 'go-')
plt.show()
```

### `'go'` $\rightarrow$ puntos verdes $\rightarrow$ gráfica de dispersión

```python
plt.plot([1,2,3,4,5], [1,2,3,4,10], 'go')
plt.show()
```

Algunos formatos usados con frecuencia son:
- `'r*--'`: **red stars with dashed lines** $\rightarrow$ **estrellas rojas con líneas discontinuas**
- `'ks.'` : **black squares with dotted line** $\rightarrow$ **cuadrados negros con línea punteada**
- `'bD-.'`: **blue diamonds with dash-dot line** $\rightarrow$ **diamantes azules con línea punto-línea**

**Para una lista completa de colores, marcadores y estilo de línea, ver el comando `help(plt.plot)` 

## Dibujar dos gráficas de dispersión el mismo *plot*

```python
#Dibujar dos conjuntos de puntos
plt.plot([1,2,3,4,5], [1,2,3,4,10], 'go')  # puntos verdes
plt.plot([1,2,3,4,5], [2,3,4,5,11], 'b*')  # estrellas azules
plt.show()
```

Ahora queremos agregar elementos básicos de una gráfica: título , leyenda, etiquetas de ejes x y y. ¿Cómo hacemos esto? Por suerte el objeto `plt` nos brinda métodos para cada uno de estos elementos. Veamos

```python
plt.plot([1,2,3,4,5], [1,2,3,4,10], 'go', label='Puntos Verdes')
plt.plot([1,2,3,4,5], [2,3,4,5,11], 'b*', label='Estrellas Azules')
plt.title('Una simple gráfica de dispersión')
plt.xlabel('eje X')
plt.ylabel('eje Y')
plt.legend(loc='best')  # legend text comes from the plot's label parameter.
plt.show()
```

## Tamaño de la figura
En nuestro jupyter notebook, esta gráfica se ve algo pequeña. Queremos incrementar el tamaño de nuestras figuras para que sean más amigables. Esto lo podemos lograr *seteando* `figsize` adentro del método `plt.figure()`.

```python
plt.figure(figsize=(10,7)) # 10 de ancho, 7 de alto
plt.plot([1,2,3,4,5], [1,2,3,4,10], 'go', label='puntos verdes') 
plt.plot([1,2,3,4,5], [2,3,4,5,11], 'b*', label='estrellas azules')
plt.title('Gráfica de dispersión')  
plt.xlabel('eje X')
plt.ylabel('eje Y')
plt.xlim(0, 6)
plt.ylim(0, 12)
plt.legend(loc='best')
plt.show()
```

Aquí introdujimos código nuevo. **¿Qué es o qué hace `plt.figure`?**

Toda gráfica (plot) que crea Matplotlib, se dibuja en un objeto llamado `figure`. Podemos pensar en `figure` como el lienzo sobre el cual vamos a dibujar todas nuestras gráficas y sus elementos.

Una figura (`figure`) puede tener uno o más `subplots` llamados `axes`, los cuales están organizados en filas y columnas. **Toda figura tiene a lo menos un `axes`**.

### NOTA: No hay que confundir el término `axes` con los ejes (llamados axis en inglés) X y Y.
