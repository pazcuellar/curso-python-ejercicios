```python
df2 = df[(df.clave_entidad == 21) | (df.clave_entidad == 12) | (df.clave_entidad == 30)]
df2 = df2[df2.subtipo_delito == 'Homicidio doloso']
df2.head()
```

```python
df2 = df2.groupby(['anio','mes','clave_entidad', 'entidad', 'subtipo_delito'])\
.agg({'frecuencia':'sum'})\
.reset_index()
df2.head()
```

```python
df2['aniomes'] = df2.apply(lambda row: int(str(row['anio'])+ str(row['mes']).zfill(2)), axis =1)
```

```python
df2.head()
```

```python
puebla_hom = df2[df2.entidad == 'Puebla']
guerrero_hom = df2[df2.entidad == 'Guerrero']
Veracruz_hom = df2[df2.entidad == 'Veracruz de Ignacio de la Llave']
puebla_hom.head()
```

```python
plt.figure(figsize=(14,7))
plt.plot(range(len(puebla_hom.aniomes)), puebla_hom.frecuencia, 'b-', label = 'Puebla')
plt.plot(range(len(guerrero_hom.aniomes)), guerrero_hom.frecuencia, 'r-', label = 'Guerrero')
plt.plot(range(len(Veracruz_hom.aniomes)), Veracruz_hom.frecuencia, 'g-', label = 'Veracruz')
plt.xticks(range(len(puebla_hom.aniomes)), puebla_hom.aniomes, rotation = 90)
plt.legend(loc='best')
plt.show()
```
