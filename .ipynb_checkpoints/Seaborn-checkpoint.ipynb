```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data_pk/pokemon.csv')
df.head()
```

```python
len(df)
```

```python
##con matplotlib
plt.plot(df.Attack, df.Defense, 'bo')
plt.xlabel('Attack')
plt.ylabel('Defense')
plt.show() #pa'quitar el texto
```

```python
#con seaborn
sns.lmplot(x='Attack', y='Defense', data=df)
plt.show() #pa' quitar el texto porque seaborn esta usando matplotlib solo
#con mejoras para visualizaci√≥n
```

```python
#con seaborn
sns.set_style('darkgrid')
sns.lmplot(x='Attack', y='Defense', data=df, fit_reg = False, hue='Type 1')
plt.show() #pa' quitar el texto porque seaborn esta usando matplotlib solo
#con mejoras para visualizaci√≥n
```

```python
sns.set_style('darkgrid')
sns.lmplot(x='Attack', y='Defense', data=df, fit_reg = False, hue='Type 1')
plt.xlim(40)
plt.xlim(20)
plt.title("Ataque y Defensa de un pokem√≥n")
plt.show() 
```

```python
plt.figure(figsize=(12,9))
sns.boxplot(data=df)
```

```python
df_stats = df.drop(['#','Total','Stage','Legendary'], axis=1)
df_stats.head()
```

```python
plt.figure(figsize=(12,9))
sns.boxplot(data=df_stats)
plt.show()
```

```python

plt.figure(figsize=(14,8))
sns.set_style('whitegrid')
sns.violinplot(x='Type 1', y='Attack', data=df)
plt.show()
```

```python

plt.figure(figsize=(14,8))
sns.set_style('darkgrid')
sns.violinplot(x='Type 1', y='Attack', data=df)
plt.show()
```
