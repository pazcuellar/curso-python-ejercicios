Una funcion que reciba un número y regresa una tupla, en la que el primer valor sea verdadero si el número es par y el segundo valor sea verdadero si el numero es mayor a 100

```python
def crear_tupla(numero):
    var1 = numero % 2 == 0
    var2 = numero > 100
    return (var1, var2)
        
 
```

```python
crear_tupla(102)
```

```python
crear_tupla(203)
```

```python
def funcion_nombre(nombre):
    var1 = nombre[0]
    longi = len(nombre)-1
    var2 = nombre[longi]
    return (var1, var2)
```

```python
f = open('paises.txt')
paises = []
for linea in f:
    diccionario_pais = {}
    linea_limpia = linea.strip()
    valores = linea_limpia.split('|')
    print(valores)
    diccionario_pais['iso'] = valores[0]
    diccionario_pais['name'] =valores[1]
    diccionario_pais['pop'] =valores[2]
    paises.append(diccionario_pais)
f.close()
```

```python
def imprimir_info_pais(pais):
    print("******{}******".format(pais['iso']))
    print("Nombre del país:", pais['name'])
    print("Población del país:", pais['pop'])
    print("*"*10)
    
peticion = ""
while peticion != "END":
    peticion = input("código del páis (END para terminar): ")
    for pais in paises:
        if peticion == pais['iso']:
            imprimir_info_pais(pais)
            
```
